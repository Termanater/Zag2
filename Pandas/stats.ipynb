```python
# Verify library versions
print("NumPy Version:", np.__version__)
print("Pandas Version:", pd.__version__)

# Load the Excel dataset
df = pd.read_excel('POP_Health_Data_Sheet_13.xlsx', engine='openpyxl')

# Debugging: Inspect the original DataFrame
print("\nOriginal DataFrame Shape:", df.shape)
print("\nOriginal DataFrame Info:")
df.info()
print("\nOriginal Null Counts:")
print(df.isnull().sum())

# Ensure Sleep_hours_weekday is numeric to avoid type issues
df['Sleep_hours_weekday'] = pd.to_numeric(df['Sleep_hours_weekday'], errors='coerce')
print("\nSleep_hours_weekday Dtype:", df['Sleep_hours_weekday'].dtype)

# Drop rows with any NaN values
df_cleaned = df.dropna()

# Debugging: Inspect the cleaned DataFrame
print("\nCleaned DataFrame Shape:", df_cleaned.shape)
print("\nCleaned DataFrame Info:")
df_cleaned.info()
print("\nCleaned Null Counts:")
print(df_cleaned.isnull().sum())

# Write the cleaned data to a new sheet in the same workbook
with pd.ExcelWriter('POP_Health_Data_Sheet_13.xlsx', engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    # Write cleaned data to a new sheet named 'Cleaned_Data'
    df_cleaned.to_excel(writer, sheet_name='Cleaned_Data', index=False)

print("\nCleaned data written to 'Cleaned_Data' sheet in 'POP_Health_Data_Sheet_13.xlsx'")
```
