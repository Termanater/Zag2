## A. Descriptive Analytics: What is happening in our campaigns?
- Compare total spend, clicks, and conversions by channel, campaign, and date.
- Examine daily trends and seasonality in traffic and conversion.
- Identify top-performing campaigns (e.g., by CTR, CVR, or ROI).

```python
#Compare total spend, clicks, and conversions by channel, campaign, and date.
# Examine daily trends and seasonality in traffic and conversion.
# Identify top-performing campaigns (e.g., by CTR, CVR, or ROI).
# Segment customers by acquisition channel, order frequency, and CLV flag
#CTR = Clicks ÷ Impressions
#CVR = Conversions ÷ Clicks
#CAC = Ad Spend ÷ New Customers

Ad_Performance_csv['CTR'] = Ad_Performance_csv['Clicks'] / Ad_Performance_csv['Impressions']
Ad_Performance_csv['CVR'] = Ad_Performance_csv['Conversions'] / Ad_Performance_csv['Clicks']
Ad_Performance_csv['CAC'] = Ad_Performance_csv['Spend_USD'] / Ad_Performance_csv['New Customers']
print(Ad_Performance_csv[['Channel', 'Campaign', 'Date', 'CTR', 'CVR', 'CAC']].head())
```
