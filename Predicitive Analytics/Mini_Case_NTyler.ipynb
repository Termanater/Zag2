```python
import pandas as pd
from pandas import get_dummies
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split

# Load the dataset
```

```python
Ad_Performance_csv = pd.read_csv('https://raw.githubusercontent.com/ttchuang/dataset/refs/heads/master/Ad_Performance.csv')
Web_Traffic_csv = pd.read_csv('https://raw.githubusercontent.com/ttchuang/dataset/refs/heads/master/Web_Traffic.csv')   
Customer_transactions_xlsx = pd.read_excel('https://raw.githubusercontent.com/ttchuang/dataset/refs/heads/master/Customer_Transactions.xlsx')

print(f'Ad Performance Data: \n {Ad_Performance_csv.head()}')
print(f' Web Traffic Data: \n {Web_Traffic_csv.head()}')       
print(f' Customer Transactions Data : \n {Customer_transactions_xlsx.head()}')    
```

## A. Descriptive Analytics: What is happening in our campaigns?
- Compare total spend, clicks, and conversions by channel, campaign, and date.
- Examine daily trends and seasonality in traffic and conversion.
- Identify top-performing campaigns (e.g., by CTR, CVR, or ROI).
- Segment customers by acquisition channel, order frequency, and CLV flag

```python
#Compare total spend, clicks, and conversions by channel, campaign, and date.
# Examine daily trends and seasonality in traffic and conversion.
# Identify top-performing campaigns (e.g., by CTR, CVR, or ROI).
# Segment customers by acquisition channel, order frequency, and CLV flag

Ad_Performance_csv.groupby(['Channel', 'Campaign_ID', 'Date'])[['Spend_USD', 'Clicks', 'Conversions']].sum()
Web_Traffic_csv.groupby('Date')[['Sessions', 'Bounce_Rate', 'Page_Load_Time_sec']].sum()

# compute order frequency per customer (there is no Order_Frequency column in the raw data)
order_freq = Customer_transactions_xlsx.groupby('Customer_ID').size().rename('Order_Frequency')

# merge the order frequency back into the customer transactions data
customer_orders = Customer_transactions_xlsx.merge(order_freq, on='Customer_ID', how='left')

# now segment by acquisition channel, order frequency, and CLV flag
customer_orders.groupby(['Acquisition_Channel', 'Order_Frequency', 'CLV_Flag'])[['Customer_ID']].nunique()
```
