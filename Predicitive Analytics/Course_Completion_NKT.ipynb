```python
#Import Packages
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns   
```

```python
#READ CSV FILE & EXPLORE DATA
Course_Completion = pd.read_csv('course_completion_data_1200.csv')
Course_Completion.head()
```

```python
# Clean Completion_Status, convert to 0/1, and show completion rate by Difficulty
Course_Completion['Completion_Status'] = (
    Course_Completion['Completion_Status']
    .astype(str)
    .str.strip()           # remove surrounding whitespace
    .str.strip("'\"")  # remove single or double quotes
)

# Hot Encoding for Completed/Incomplete
Course_Completion['Completion_Status'] = Course_Completion['Completion_Status'].map({'Completed': 1, 'Incomplete': 0})

# Completion rate by Difficulty
completion_by_diff = (
    Course_Completion.groupby('Difficulty_Level')['Completion_Status']
    .agg(['mean', 'count'])
    .rename(columns={'mean': 'completion_rate'})
)
completion_by_diff['completion_rate'] = completion_by_diff['completion_rate'].round(3)

completion_by_diff
```
