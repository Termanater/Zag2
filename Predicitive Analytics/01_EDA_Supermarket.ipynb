# Split-Apply-Combine

```python
# -- What is the total amount of Different Customer Types? --
df.groupby('Customer type')['Total'].sum()
```

```python
# -- What is the average amount of Different Customer types? --
df.groupby('Customer type')['Total'].mean()
```

```python
# -- Create a cross table Customer Type by Gender showing their mean totals -- 

df.groupby(['Customer type','Gender'])['Total'].mean().unstack(level=1)
```
